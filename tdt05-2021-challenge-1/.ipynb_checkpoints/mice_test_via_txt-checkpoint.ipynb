### Note: use read.table and write.table instead of read.csv and write.csv.

```python
import os
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data = pd.read_csv('train.csv')
columns = list(data.columns)
data
```

```python
infile = 'train_missing.csv'
outfile = 'train_imputed.csv'
```

```python
np.savetxt(infile, data.to_numpy(), delimiter=",")
os.system("Rscript --vanilla run_mice.R %s %s" % (infile, outfile))
data_imputed = np.genfromtxt(outfile, delimiter=",")
```
