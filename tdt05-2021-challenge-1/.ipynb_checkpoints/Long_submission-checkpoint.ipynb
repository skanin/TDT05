## Types of features

There exists multiple types of features, eg. Ordinal features, Categorical and Continous.

### Categorical features
Categorical features has one or more _categories_ eg., "sex" (male, female, non-binary). These features cannot be sorted.

To be able to do see correlations and other data analysis on the categorical variables, we will first do some preprocessing on them (encode them as floats)

The first thing we notice is that **f1** can be country codes, so we do nothing with them just now (AU = Australia, JP = Japan , IT = Italy ...). **f11** and **f9** however, are the only ones that have two letters in them. We will split these in new columns

```python
new_col_names = []

def split_cols(df : pd.DataFrame, cols_to_split : list) -> None:
    for i in range(1, len(cols_to_split) + 1):
        new_col_names.append(cols_to_split[i-1] + f'_{i}')
        
    def add_space(x : str):
        if type(x) != str:
            return x
        
        st = ""
        for i, ch in enumerate(x):
            st += ch + (" " * int(i != (len(x) - 1)))
        return st
    
    for col in cols_to_split:
        n = len(df[col][0])
        df[col] = df[col].apply(lambda x: add_space(x))
        df[[col + f'_{i+1}' for i in range(n)]] = df['f9'].str.split(' ', 1, expand=True)
    
    df.drop(columns=cols_to_split, inplace = True)

split_cols(X, ['f9', 'f11'])
split_cols(X_test, ['f9', 'f11'])

other_features = list(set(X.columns).intersection(other_features).union(set(new_col_names)))

X[other_features].head()
```

Now, we can encode them easily with pandas category type

```python
cols = other_features

for col in cols:
    X[col] = X[col].astype('category')
    X[col] = X[col].cat.codes
    
    X_test[col] = X_test[col].astype('category')
    X_test[col] = X_test[col].cat.codes
    
X[other_features].head()
```

Further, we have to encode the hex columns. We choose to convert them directly to floats.

```python
def conv_hex(df : pd.DataFrame) -> None:
    def conv_hex_map(x):
        try:
            return int(x, 16)
        except ValueError as e:
            return np.nan
        except TypeError as e:
            return np.nan

    for col in hex_features:
        if col not in list(df.columns):
            continue
        col_loc = list(df.columns).index(col)
        df[col] = df[col].apply(lambda x: conv_hex_map(x))

conv_hex(X)
conv_hex(X_test)
        
X[hex_features].head()
```

# Selecting the model

First, we define some functions for splitting, fitting and predicting

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split

def split(X : pd.DataFrame, y : pd.DataFrame, size=.2):
    return train_test_split(X, y, test_size=size, stratify=y)

def fit(data : pd.DataFrame, labels : pd.DataFrame, classifier : any, eval_pool = None) -> None:
    if eval_pool is not None:
        classifier.fit(data, labels, eval_set=eval_pool)
    else:
        classifier.fit(data, labels)

def predict_proba(classifier : any, test : pd.DataFrame) -> np.array:
    return classifier.predict_proba(test)

def predict(classifier : any, test : pd.DataFrame) -> np.array:
    return classifier.predict(test)

def print_accuracy(pred : np.array, test : pd.DataFrame, name : str):
    print(name + ' Model AOC score: {0:0.4f}'.format(roc_auc_score(test, pred[:, 1])))
    
def real_predict(clf, name, submission=True):
    prediction = predict_proba(clf, X_test_split)

    print_accuracy(prediction, y_test_split, name)

    if submission:
        real_pred = predict_proba(clf, X_test)

        f = open("./pred.csv", "w")
        f.write("id,target\n")
        id_nr = 50000
        for v in real_pred[:, 1]:
            f.write(f"{id_nr},{v}\n")
            id_nr += 1
        f.close()
```

### GridSearchCV
Then, we set up a function for doing **GridSearchCV**
_GridSearchCV_ is a function from the model_selection package in Sklearn that helps find the best parameters for the estimator. It loops through predefined hyperparameters and fits the model training on the training set. After executing it, by the end, it has a list of the best hyperparameters for the model/estimator.

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

def grid(estimator, params, n_jobs=-1):
    grid_cv = GridSearchCV(estimator=estimator, param_grid=params, cv=5, n_jobs=n_jobs, scoring='roc_auc', refit=True, verbose=10)
    grid_cv.fit(X_train, y)
    return grid_cv
```

## Stacking them all together

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier

n = len(X_train)

class_0 = len(X_train_pre[X_train_pre['target'] == 0])
class_1 = len(X_train_pre[X_train_pre['target'] == 1])

weights=[class_0/n, class_1/n]

cat_clf_params = {
    'depth': 1, 
    'iterations': 2000, 
    'l2_leaf_reg': 1e-20, 
    'leaf_estimation_iterations': 10, 
    'loss_function': 'CrossEntropy',
    'boosting_type': 'Plain', 
    'gpu_cat_features_storage': 'CpuPinnedMemory',
    'max_ctr_complexity':1,
    'learning_rate':1, 
    'verbose':False, 
    'random_state':42, 
}

lgbm_clf_params = {
    'boosting_type': 'dart', 
    'colsample_bytree': 0.64, 
    'learning_rate': 1, 
    'max_bin': 255, 
    'n_estimators': 32, 
    'num_leaves': 16, 
    'objective': 'binary', 
    'subsample': 0.7,
    'random_state':42, 
    'device':'gpu',
    'scale_pos_weight': class_0/class_1
}

xgb_clf_params = {
    'colsample_bytree': 0.6,
    'gamma': 0.5,
    'learning_rate': 0.1,
    'max_depth': 3,
    'min_child_weight': 5,
    'n_estimators': 1000,
    'subsample': 1.0,
    'tree_method':'gpu_hist',
    'gpu_id':0, 
    'verbosity': 0, 
    'objective':'binary:logistic', 
    'silent':True,
    'scale_pos_weight': class_0/class_1
}

ada_clf_params = {
    'learning_rate': 1, 
    'n_estimators': 2000
}

rf_clf_params = {
    'criterion': 'entropy', 
    'max_depth': 8, 
    'max_features': 'log2', 
    'n_estimators': 500
}

lr_clf_params = {
    'C': 0.1, 
    'class_weight': {0: 0.81272, 1: 0.18728}, 
    'max_iter': 10000, 
    'penalty': 'l2', 
    'solver': 'saga'}

stacking_models = {}

stacking_models['cat'] = CatBoostClassifier(**cat_clf_params)
stacking_models['lgbm'] = LGBMClassifier(**lgbm_clf_params)
stacking_models['xgb'] = XGBClassifier(**xgb_clf_params)
stacking_models['ada'] = AdaBoostClassifier(**ada_clf_params)
stacking_models['rf'] = RandomForestClassifier(**rf_clf_params)

clf = StackingClassifier(estimators = list(stacking_models.items()), final_estimator=LogisticRegression(**lr_clf_params), cv=5, n_jobs=-1, stack_method="predict_proba")
fit(X_train, y, clf)
```
